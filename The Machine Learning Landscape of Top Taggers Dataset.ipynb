```python
n=np.zeros(shape=200)
phi=np.zeros(shape=200)
En=np.zeros(shape=200)
```

```python
j=1
for i in range(200):
    df1 = df.iloc[j:j+1,[0+4*i,1+4*i,2+4*i,3+4*i]]
   # print(df1)
    Px=df1.iloc[0][1]
    Py=df1.iloc[0][2]
    Pz=df1.iloc[0][3]
    E=df1.iloc[0][0]
    P=np.sqrt(Px**2+Py**2+Pz**2)
    if (P!=0):
        theta=np.arccos(Pz/P)
        n[i]=-np.log(np.tan(theta/2))
    else:
        n[i]=np.nan
    if (Px!=0):
        phi[i]=np.arctan(Py/Px)
    else:
        phi[i]=np.nan
    if (E!=0):
        En[i]=E
    else:
        En[i]=np.nan
   # print("Rapidity:", n[i])
   # print("Azimutal:", phi[i])
   # print("Energy:", En[i])
    
```
